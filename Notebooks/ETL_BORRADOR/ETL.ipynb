```python
# Eliminar filas duplicadas basadas en 'identificacion' y 'nombre_completo'
df = df.drop_duplicates(subset=['identificacion', 'nombre_completo'])
```

```python
# Contar cuántos registros tienen el signo de interrogación
contar_interrogaciones = df['nombre_completo'].str.contains(r'\?').sum()
```

```python
# Función para corregir palabras
def corregir_palabras(texto):
    if isinstance(texto, str):
        return str(TextBlob(texto).correct())
    return texto

# Aplicar la corrección a una columna específica (ejemplo: 'nombre_columna')
df['nombre_completo'] = df['nombre_completo'].apply(corregir_palabras)
```
