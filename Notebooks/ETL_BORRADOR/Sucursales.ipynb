```python
import pandas as pd
```

```python
df_sucursales = pd.read_excel('healthtech.xlsx', sheet_name ='DimSucursales')
```

```python
# Convertir los nombres de columnas a minúsculas
df.columns = df.columns.str.lower()
```

```python
# Verificar si hay valores nulos en cada columna
nulos_totales = df.isnull().sum()

# Verificar si hay valores vacíos (cadenas vacías) en las columnas del DataFrame
vacios_totales = (df == '').sum()

# Mostrar los resultados
print("Valores nulos por columna:\n", nulos_totales)
print("\nValores vacíos por columna:\n", vacios_totales)


# Verificar si hay filas duplicadas en todo el DataFrame
duplicados = df.duplicated()

# Contar cuántas filas duplicadas hay
numero_duplicados = duplicados.sum()

# Mostrar el número de duplicados
print(f"\nNúmero de filas duplicadas: {numero_duplicados}")
```

```python
# Renombrar la columna 'codigo_sucursal' a 'id_sucursal'
df = df.rename(columns={'codigo_sucursal': 'id_sucursal'})
```

```python
# Reemplazar los valores en la columna 'id_sucursal'
df['id_sucursal'] = df['id_sucursal'].replace({1: 1022, 2: 1023})
```
